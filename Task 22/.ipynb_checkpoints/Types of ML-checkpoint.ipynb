# Supervised Machine Learning
Supervised learning is a type of machine learning where the model is trained on labeled data. This means that the data has inputs (also known as features) and corresponding outputs (also known as labels or targets) that the model is trying to predict. The goal of supervised learning is to learn a function that maps inputs to outputs.

## Types of supervised learning
- Classification
- Regression.

### Classification
A type of supervised learning, where the algorithm is trained on labeled data and used to predict the class or category of new, unlabeled data. The goal of classification is to build a model that accurately predicts the class of an unknown data point based on its features.<br>
Here are some popular classification algorithms:
- Logistic Regression
- Naive Bayes
- k-Nearest Neighbors (k-NN)
- Decision Tree
- Random Forest
- Support Vector Machine (SVM)
- Gradient Boosting
- Neural Networks

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Load the iris dataset
iris = datasets.load_iris()

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)

# Create a k-NN classifier with k=3
knn = KNeighborsClassifier(n_neighbors=3)

# Train the classifier on the training data
knn.fit(X_train, y_train)

# Predict the classes of the test data
y_pred = knn.predict(X_test)

# Print the accuracy of the classifier
print("Accuracy:", knn.score(X_test, y_test))
```

### Regression
A type of supervised learning algorithm used to predict a continuous output variable based on one or more input variables. It is used for modeling the relationship between the target variable and the independent variables.<br>
Some popular regression algorithms
- Linear Regression
- Decision Tree Regression
- Support Vector Regression
- Random Forest Regression
- Polynomial Regression
- Elastic Net Regression
- K-Nearest Neighbors Regression
- Neural Network Regression

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Load the diabetes dataset
diabetes = load_diabetes()
X = diabetes.data[:, np.newaxis, 2]
y = diabetes.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Fit the model
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predict the test set results
y_pred = regressor.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R2 Score:", r2)
```

# Unsupervised learning
A type of machine learning in which the model learns patterns and relationships in the data without being explicitly trained on labeled data. This means that the data is unstructured and the algorithm must find its own structure. The goal of unsupervised learning is to discover underlying patterns or groupings in the data.

## Types of unsupervised learning
- Clustering
- Dimensionality Reduction

### Clustering
An unsupervised learning technique in which similar instances are grouped together based on the similarity of their attributes or characteristics. In other words, clustering is the process of partitioning a set of data points into groups, called clusters, such that data points in the same cluster are more similar to each other than to those in other clusters. <br>
Some popular clustering algorithms:
- K-Means clustering
- Hierarchical clustering (Agglomerative and Divisive)
- Density-based clustering (DBSCAN)
- Mean-shift clustering
- Fuzzy clustering (FCM)

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Fit KMeans clustering algorithm to the dataset
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)

# Visualize the clustering results
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title('KMeans Clustering Results')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

### Dimensionality Reduction
Dimensionality Reduction is a process of reducing the number of features in a dataset while retaining most of the relevant information. It is used to tackle the problem of high-dimensional data, which can be computationally expensive and difficult to visualize<br>
There are two types of dimensionality reduction techniques:
1. <b>Feature Selection</b>: In this approach, a subset of the original features is selected and the remaining features are discarded. Feature selection techniques include correlation-based feature selection, recursive feature elimination, and principle component analysis.
2. <b>Feature Extraction</b>: In this approach, new features are created from the original features. 

Some popular algorithms used in dimensionality reduction
- Principal Component Analysis (PCA)
- Linear Discriminant Analysis (LDA)
- Non-negative Matrix Factorization (NMF)
- Independent Component Analysis (ICA)
- t-Distributed Stochastic Neighbor Embedding (t-SNE)
- Autoencoder

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Reading the Data
data = load_iris()
X = data.data
y = data.target

# Fitting the Model
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Visualizing the Data in Two Dimensions
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
plt.title('Iris Data in Two Dimensions')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```

# Self-supervised learning

A type of machine learning in which a model learns to predict a specific property or feature of the input data without explicit supervision. Unlike supervised learning, where labeled data is used to train a model, self-supervised learning algorithms use the input data itself to create the labels.

## Types of self-supervised learning:

1. <b>Autoencoder</b>: Autoencoders are neural networks that are trained to reconstruct input data. They consist of two parts: an encoder that maps the input data to a compressed representation, and a decoder that maps the compressed representation back to the original data. The model is trained by minimizing the difference between the original input data and the reconstructed output.

2. <b>Contrastive Learning</b>: In contrastive learning, the model learns to distinguish between similar and dissimilar inputs. It learns to identify pairs of inputs that are similar and pairs that are dissimilar. The model is trained by minimizing the distance between similar pairs and maximizing the distance between dissimilar pairs.

3. <b>Generative Models</b>: Generative models learn to generate new samples that are similar to the input data. The model is trained by maximizing the likelihood of the input data given the model parameters. Variational Autoencoders (VAE) and Generative Adversarial Networks (GANs) are examples of generative models.

4. <b>Multi-Task Learning</b>: In multi-task learning, the model is trained to perform multiple tasks simultaneously. The tasks are related to each other and the model learns to use the information from one task to improve the performance of another task.

# Reinforcement Learning 
A type of machine learning that involves training an agent to take actions in an environment to maximize a cumulative reward. The agent learns through trial and error, receiving feedback in the form of positive or negative rewards for each action it takes. The goal of reinforcement learning is to find the optimal policy or sequence of actions that maximizes the expected cumulative reward.

## Types of reinforcement learning
1. Model-based RL: In model-based RL, the agent learns a model of the environment which is used to plan future actions. The agent uses this model to predict the next state and reward given the current state and action. Then, it selects the action that maximizes the expected cumulative reward based on the predicted outcomes.

2. Model-free RL: In model-free RL, the agent directly learns the optimal policy without building a model of the environment. The agent uses trial and error to learn the best actions to take in each state to maximize the expected cumulative reward.

3. Value-based RL: In value-based RL, the agent learns the optimal value function which estimates the expected cumulative reward of taking an action in a given state. The agent selects the action that maximizes the value function.

4. Policy-based RL: In policy-based RL, the agent learns the optimal policy directly. The agent uses a function approximator to estimate the policy, which is a mapping from states to actions. The function approximator is trained to maximize the expected cumulative reward.

5. Actor-critic RL: In actor-critic RL, the agent learns both a policy and a value function. The actor is responsible for selecting actions based on the policy, and the critic evaluates the policy by estimating the value function.

6. Multi-agent RL: In multi-agent RL, there are multiple agents that learn to cooperate or compete with each other to maximize the expected cumulative reward.

7. Inverse RL: In inverse RL, the agent learns the reward function instead of the optimal policy. The reward function is learned from expert demonstrations, and then used to learn the optimal policy.

8. Meta-RL: In meta-RL, the agent learns how to learn. The agent learns a meta-policy that can adapt to new environments more quickly.
